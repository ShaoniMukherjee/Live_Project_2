```python
def word_divide_char(inputs):
    characters=[]
    for i in inputs:
        characters.append(i)
    return characters
```

```python
word_divide_char("kzde5577")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer(tokenizer=word_divide_char)
```

```python
X=vectorizer.fit_transform(X)
```

```python
## Logistics Regression

from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state=42) #splitting
```

```python
log_class = LogisticRegression(penalty='l2',multi_class='ovr')
log_class.fit(X_train,y_train)
```

```python
log_class.intercept_
```

```python
log_class.coef_
```

```python
y_pred = log_class.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
confusion_matrix(y_test, y_pred)
```

```python
log_class.score(X_test, y_test)
```

```python
print(classification_report(y_test, y_pred))
```
